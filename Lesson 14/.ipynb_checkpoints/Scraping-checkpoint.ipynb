```python
import requests
```

```python
URL = 'https://habr.com/ru/all'
KEYWORDS = ['года', 'парсинг']
```

```python
res = requests.get(URL)
```

```python
res.text
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

```python
page_content = soup.find('div', class_='posts_list')  
page_content 
```

```python
article_blocks = page_content.find_all('article', class_='post_preview')
article_blocks
```

```python
import pandas as pd
```

```python
KEYWORDS = ['python', 'парсинг']
habr_news = pd.DataFrame()

for block in article_blocks:
    #preview =  lambda x: x.find('div', class_='post__text').text
    #preview = preview.lower().split()
    preview = block.find('div', class_='post__text').text.strip().lower().split()
    new_list = list(set(KEYWORDS) & set(preview))
    if len(new_list) == 0:
        continue
    link = block.find('a', class_= 'post__title_link').get('href')
    title = block.find('a', class_= 'post__title_link').text
    soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    date = pd.to_datetime(soup.find('span', class_='post__time').get('data-time_published'), dayfirst=True).date()
    row = {'date': date, 'title': title, 'link': link}
    habr_news = pd.concat([habr_news, pd.DataFrame([row])])  
habr_news
```
