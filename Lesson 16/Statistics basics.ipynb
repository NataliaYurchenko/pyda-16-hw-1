Сильный разбег в дыхании - параметром можно пренебречь.
В возрасте должно быть всего 2 значения, остальное - ошибки, которые нужно исправить (выбросы заменить).
Температура и пульс будут ключевыми понятиями. В температуре много пропусков, можно заменить значением моды (хотя различие между тремя средними незначительное).
Температура конечностей - качественный показатель, много пропусков. Заменим модой. То же самое касается периферийного пульса.


```python
q1_1 = df.Age.quantile(0.25)
q3_3 = df.Age.quantile(0.75)
iqr_1 = q3_3 - q1_1
lower_bound_1 = q1_1 - (1.5 *iqr_1)
upper_bound_1 = q3_3 + (1.5 *iqr_1)
remove_outliers_age = df[df.Age.between(lower_bound_1, upper_bound_1, inclusive = True)].sort_values('rectal temperature')
remove_outliers_age_1 = remove_outliers_age[['Age', 'rectal temperature']]
remove_outliers_age_1
```

```python
df_age = df[['Age', 'rectal temperature']]
df_age
```

```python
pd.concat([df_age, remove_outliers_age_1]).drop_duplicates(keep = False)
```

выброс в возрасте - точно ошибка. Надо заменить модой и выбросы, и пропуски

```python
age_filled_mode = df['Age'].fillna(df['Age'].mode())
age_filled_mode
```

заменили пропуски на моду

```python
q1 = df.pulse.quantile(0.25)
q3 = df.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound= q1 - (1.5 *iqr)
upper_bound= q3 + (1.5 *iqr)
remove_outliers = df[df.pulse.between(lower_bound, upper_bound, inclusive = True)].sort_values('rectal temperature')
remove_outliers_1 = remove_outliers[['pulse', 'rectal temperature']]
remove_outliers_1
```

```python
df_1 = df[['pulse', 'rectal temperature']]
df_1
```

```python
pd.concat([df_1, remove_outliers_1]).drop_duplicates(keep = False)
```
