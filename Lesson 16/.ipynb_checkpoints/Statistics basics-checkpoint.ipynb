Сильный разбег в дыхании - параметром можно пренебречь.
В возрасте должно быть всего 2 значения, 1 и 9. 9 (предположительно) ошибочные, и вместо них должно быть хначение 2.
Температура и пульс будут ключевыми понятиями. В температуре много пропусков, можно заменить значением моды (хотя различие между тремя средними незначительное).
Температура конечностей - качественный показатель, много пропусков. Заменим модой. То же самое касается периферийного пульса.


```python
q1_1 = df.Age.quantile(0.25)
q3_3 = df.Age.quantile(0.75)
iqr_1 = q3_3 - q1_1
lower_bound_1 = q1_1 - (1.5 *iqr_1)
upper_bound_1 = q3_3 + (1.5 *iqr_1)
```

выбросов в возрасте нет.

Пропуски в возрасте нужно заменить модой. Значение 9 (ошибочное) меняем на 2.


```python
df['Age'] = df['Age'].fillna(df['Age'].mode()).replace(9,2)

df
```

заменили пропуски на моду, а также исправили ошибку.

```python
for col in df.columns:
    pct_missing = df[col].isna().mean()
    print (f'{col} - {pct_missing :.1%}')
```

Выбросы не удаляем, так как они носят естественный характер. Заменим пропуски для важных колонок (пульс, температура). Проверим, отличается ли средняя температура для разных возрастных групп.

```python
df.groupby('Age').mean().head()
```

Средняя температура не сильно зависит от возраста. Заполним пропуски.

```python
rectal_temperature_filled_mean = df['rectal temperature'].fillna(df['rectal temperature'].mean())
rectal_temperature_filled_mean
```

```python

print(rectal_temperature_filled_mean.mean())
print(rectal_temperature_filled_mean.mode()[0])
print(rectal_temperature_filled_mean.median())
print(rectal_temperature_filled_mean.std())
print(rectal_temperature_filled_mean.var())
```

около 19% значений температуры были пропущены. Это важный показатель, поэтому его нужно сохранить. Заменим на среднее арифметическое, хотя изначальные значения моды, медианы и ср арифметич были очень близки (можно было выбрать любое).

частота дыхания также имеет большое количество пропусков, но колебания этого показателя достаточно сильные, а ценность не высокая. 

Среднее значение для пульса сильно отличается для разных возрастных групп. Найдем медиану для каждой группы и заменим пропуски медианными значениями.

```python
df.loc[df['Age']== 1, 'pulse'] = df['pulse'].fillna(df[df['Age']== 1]['pulse'].median())
df.loc[df['Age']== 2, 'pulse'] = df['pulse'].fillna(df[df['Age']== 2]['pulse'].median())
df
```

```python
for col in df.columns:
    pct_missing = df[col].isna().mean()
    print (f'{col} - {pct_missing :.1%}')
```
