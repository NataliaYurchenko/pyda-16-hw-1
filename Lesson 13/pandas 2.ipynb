```python
ratings = pd.read_csv(r'C:\Users\vj0467\OneDrive - Sandvik\HomeShare\Documents\Yurchenko desktop copy\Marketing\Нетология_2020\HW\pyda-16-hw-1\ml-latest-small\ratings.csv')
```

```python
ratings.head()
```

```python
ratings_1 = ratings.groupby('userId').count()
ratings_1.head()
```

```python
ratings_1 = ratings_1[ratings_1['movieId'] > 100]
```

```python
results = ratings.groupby('userId').agg(['min', 'max'])['timestamp']
```

```python
results.head()
```

```python
def average_LT (row):
    return row['max'] - row['min']
```

```python
results['average_LT'] = results.apply(average_LT, axis=1)
results.head()
```

```python

final = ratings_1.merge(results, on = 'userId', how = 'left')

final[['movieId','average_LT']].head()
```
